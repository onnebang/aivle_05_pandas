```python
bin = list(range(0,91,10))
bin
```

```python
label = []
for b in bin :
    if b == 0 :
        name = '유아'
    else :
        name = str(b) + '대'
    label.append(name)
label
```

```python
bin = list(range(0,91,10))
label = []
for b in bin[:9] :
    if b == 0 :
        name = '유아'
    else :
        name = str(b) + '대'
    label.append(name)

pd.cut(data['Age'], bins = bin, labels = label, right = False )
```

### 실습 5 : Group by

* 타이타닉 데이터를 이용하시오

* Q1. 객실 등급별 평균 나이

```python
data.groupby('Pclass', as_index=False)['Age'].mean()
```

* Q2. 승선 지역별 평균 운임

```python
data.groupby('Embarked', as_index=False)['Age'].mean()
```

* Q3. 승선 지역별 생존 여부별 탑승객 수

```python
data.groupby(['Embarked','Survived'], as_index=False).count()
```

```python
tmp = data.groupby(['Pclass','Survived'], as_index=False)['Fare']
tmp
```
